```python
sql_query1 = 'SELECT householdid, productid, storeid, weeknum, daynum, transactiontime FROM transactiondata_subset WHERE householdid BETWEEN 601 AND 800;'
df1 = pd.read_sql(sql_query1, engine)
df1.describe(include = "all")
```

```python
def convert_transac_time(time):
    time = str(time).strip()
    if len(time) < 4:
        time = time.zfill(4)
    hour = int(time[:2])
    min = int(time[2:4])
    return f'{hour:02d}:{min:02d}:00'

df1_1_1.loc[:,'transactiontime'] = df1_1_1['transactiontime'].apply(convert_transac_time)
df1_1_1.loc[:,'transactiontime'] = pd.to_datetime(df1_1_1['transactiontime'], format = "%H:%M:%S")
```
